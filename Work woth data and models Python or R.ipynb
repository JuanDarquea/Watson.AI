```python
# Bulleted list examples
bulleted_list_2 = [
    "Data analysis and visualization",
    "Machine learning model development", 
    "Statistical analysis and reporting"
]
print("BULLETED LIST")
for item in bulleted_list_2:
    print(f"• {item}")
```

```python
# Numbered list examples - Using Python list with enumerate or dict
numbered_list_2 = [
    "Collect and clean data",
    "Perform exploratory data analysis",
    "Build and train models"
]

# Alternative: Using dictionary for numbered list
numbered_dict = {
    "Collect and clean data",
    "Perform exploratory data analysis", 
    "Build and train models"
}

print("\nNUMBERED LIST")
for i, item in enumerate(numbered_list_2, 1):
    print(f"{i}. {item}")
    
print("\nNUMBERED DICT")
for i, item in enumerate(numbered_dict, 1):
    print(f"{i}. {item}")
```

```python
# Table examples - Using list of dictionaries (most common approach)
table_data = [
    {"Name": "Alice", "Role": "Data Scientist", "Experience": "5 years"},
    {"Name": "Bob", "Role": "ML Engineer", "Experience": "3 years"},
    {"Name": "Carol", "Role": "Data Analyst", "Experience": "2 years"}
]
print("\nTABLE (Dictionary approach)")
for row in table_data:
    print(f"Name: {row['Name']}, Role: {row['Role']}, Experience: {row['Experience']}")

# Alternative: Using list of lists for table
table_matrix = [
    ["Name", "Role", "Experience"],  # Header
    ["Alice", "Data Scientist", "5 years"],
    ["Bob", "ML Engineer", "3 years"],
    ["Carol", "Data Analyst", "2 years"]
]
print("\nTABLE (Matrix approach)")
headers = table_matrix[0]
for row in table_matrix[1:]:
    for i, cell in enumerate(row):
        print(f"{headers[i]}: {cell}", end=" | ")
    print()  # New line after each row

# Alternative: Using pandas DataFrame (if you have pandas)
import pandas as pd

table_df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Carol"],
    "Role": ["Data Scientist", "ML Engineer", "Data Analyst"],
    "Experience": ["5 years", "3 years", "2 years"]
})
print("\nTABLE (Pandas DataFrame)")
print(table_df,)
```

```python
print("\n=== ACCESSING DATA ===")
print(f"First bulleted item: {bulleted_list_2[1]}")
print(f"Second numbered item: {numbered_list_2[1]}")
print(f"Alice's role: {table_data[0]['Role']}")
print(f"Bob's experience: {table_df.loc[1, 'Experience']}")
```

```python
# MODIFYING THE DATA STRUCTURES
print("\n=== MODIFYING DATA ===")
# Add to bulleted list
bulleted_list_2.append("Data visualization dashboards")
print(f"Updated bulleted list: {bulleted_list_2}")
```
